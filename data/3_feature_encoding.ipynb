### Encode Nominal Variables

```python
nominal_variables = ['gender','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','PaymentMethod']
```

```python
# Nominal variables
nominal_variables = [
    'gender',
    'Partner',
    'Dependents',
    'PhoneService',
    'MultipleLines',
    'InternetService',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'PaperlessBilling',
    'PaymentMethod'
]

# Create dummies in place
for col in nominal_variables:
    dummies = pd.get_dummies(df[col], prefix=col)
    df = pd.concat([df, dummies], axis=1)
    del df[col]

df.head()
```

### Encord Ordinal Variables

```python
# Ordinal encoding mapping for 'Contract'
encode_dict_contract = {
    'Month-to-month': 0,
    'One year': 1,
    'Two year': 2
}

# SeniorCitizen is already numeric (0/1), so we don't need to map unless it's 'Yes'/'No'
if df['SeniorCitizen'].dtype == 'object':
    encode_dict_senior = {'No': 0, 'Yes': 1}
    df['SeniorCitizen'] = df['SeniorCitizen'].map(encode_dict_senior)

# Map Contract
df['Contract'] = df['Contract'].map(encode_dict_contract)

df.head()
```
